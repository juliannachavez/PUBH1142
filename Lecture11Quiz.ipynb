## Question 8


### Create an array of 1000 bootstrapped sample means (using the sample size of 40) and assign the array to the variable bootstrapped_means and sort the values in the array in ascending order.





```python
sample_bootstrap_vals = [numpy.random.choice(sample, 40, replace=True).mean() for _ in range(1000)]

```

```python
numpy.random.normal()
```

```python

fig = graph_objects.Figure(
    graph_objects.Histogram(
        x=sample_bootstrap_vals,
        histnorm='probability'
    )
)

fig.update_layout(
    title='Histogram of bootstrapped means',
    xaxis_title='Bootstrap sample means',
    yaxis_title='Probability',
    width=700
)

fig.show()

```

```python
bootstrap_diffAB = [numpy.random.choice(sample, 40, replace=True).mean() - numpy.random.choice(sample, 40, replace=True).mean() for _ in range(1000)]
bootstrap_diffAB.sort()



```
